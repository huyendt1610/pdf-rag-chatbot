```python

import os
from dotenv import load_dotenv
load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain.agents import AgentType, initialize_agent, load_tools


google_api_key = os.getenv("GOOGLE_GEMINI_API_KEY")

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.6, google_api_key=google_api_key)

tools = load_tools(["wikipedia"], llm=llm)
# tools = load_tools(["serpapi", "llm-math"], llm=llm)

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, # talk & action
    verbose=True
)

agent.run("What was the height of the tallest building in the world in 2010, and what is its current height?")
```
