```python
from sentence_transformers import SentenceTransformer # pip install ipywidgets 
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("HUGGINGFACEHUB_API_TOKEN")
```

```python
encoder = SentenceTransformer("all-mpnet-base-v2", token=api_key)
encoder.encode
```

```python
vectors = encoder.encode(df.text)
vectors.shape # embedding dim of transformer models is usually 768, but it can be different depending on the model used
```

```python
dim = vectors.shape[1]
```

```python
import faiss
```

```python
index = faiss.IndexFlatL2(dim) # index will use L2 distance to calculate the similarity between vectors, but there are other options such as inner product (IP) or cosine similarity (COSINE) depending on the use case. L2 is commonly used for dense vector search, while IP and COSINE are often used for sparse vector search. 
index.reset() # or rest first 
index.add(vectors) # need to run the above cell to reset the index before adding to avoid duplicates
print(index.ntotal) # number of vectors in the index
```

```python
search_query = "I want to buy a polo t-shirt"
vec = encoder.encode(search_query)
vec.shape
```

```python
vec = vec.reshape(1,-1) 
vec.shape
```

```python
# return the top 2 most similar vectors to the search query, along with their distances. The smaller the distance, the more similar the vectors are.
# The output will be a tuple of two arrays: the first array contains the distances of the nearest neighbors, and the second array contains the indices of the nearest neighbors in the original dataset.
distances, I = index.search(vec, k=2) 
```
